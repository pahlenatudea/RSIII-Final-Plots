```python
import numpy as np
import pandas as pd
```

```python
z=loadtxt('dataLHC/chargino_pairprod.txt')
y60=np.loadtxt('dataLHC/data_stau60.txt')
y80=np.loadtxt('dataLHC/data_stau80.txt')
y40=np.loadtxt('dataLHC/data_stau40.txt')
```

```python
columns1=['m_stau','xs_exp','xs_obs']
columns2=['m_ch','xs','delta_xs']
yd=pd.DataFrame(y60,columns=columns1)
zd=pd.DataFrame(z,columns=columns2)
yd80=pd.DataFrame(y80,columns=columns1)
yd40=pd.DataFrame(y40,columns=columns1)
```

```python
plt.semilogy(zd['m_ch'],zd['xs'],'b-',lw=1.5,label= r'$ \sigma( p p \to \Sigma^{+}_1 \Sigma^{-}_1)_{\rm{NLO}} $')
plt.semilogy(yd['m_stau'],yd['xs_obs'],'ko',label='')
plt.semilogy(yd80['m_stau'],yd80['xs_obs'],'wo',label='')
plt.semilogy(yd['m_stau'],yd['xs_obs'],'r-', lw=2,  label= r'$m_{H^0}=60\ {\rm GeV} $')
plt.semilogy(yd80['m_stau'],yd80['xs_obs'],'r-', lw=2, color='#ffbb00', label= r'$m_{H^0}=80\ {\rm GeV} $')
plt.xlabel(r'$  m_{\Sigma^{\pm}_1}  {\rm [GeV]}$',size=15)
plt.ylabel(r'$ \sigma[{\rm pb}]$',size=15)

plt.legend(loc='best')
plt.legend(prop={'size':14},loc='best')
#axhline(y=1.43,color='black')
plt.axvline(103.5,color = 'gray',linewidth=1.,linestyle='--') # M_S=300.0
plt.annotate('LEP excluded',fontsize=12, xy=(85, 0.4), xytext=(90, .75), color='black',rotation=90)
#plt.annotate('LEP excluded (winos)',fontsize=12, xy=(85, 0.1), xytext=(93, .4), color='black',rotation=90)
#plt.axvline(95.,color = 'green',linewidth=1.,linestyle=':') # M_S=300.0
#plt.annotate('(staus)',fontsize=12, xy=(85, 0.1), xytext=(83, 2.), color='green',rotation=90)
plt.annotate('${\mathcal{B}}_{\\tau}=1$',fontsize=12, xy=(300, 0.4), xytext=(320, .018), color='black')
#figsize(5,4)
figsize(5.5,4.5)
plt.xlim(80,400)
#plt.xlim(600,700)
plt.ylim(1E-2,1E1)
#plt.ylim(0.01,10000)
#plt.grid(color='k', linestyle='--', linewidth=0.5)
matplotlib.rcParams.update({'font.size': 10})
plt.tight_layout(rect=(0.05,0.05,1,1))
#plt.title('LHC $95\%$ CL excluded, ${\mathcal{B}}_{\\tau}=1$')
#plt.savefig('tauphilexclge.pdf')
```
